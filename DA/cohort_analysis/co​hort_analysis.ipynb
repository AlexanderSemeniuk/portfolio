```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import datetime as dt 
```

```python
data=pd.read_excel('26.4_когортный анализ.xlsx')
```

```python
data.head(20)
```

```python
data.user_id.value_counts ()
```

```python
data.info()
```

```python
def get_month(x):
    return dt.datetime(x.year,x.month,1)
```

```python
data['month_year']=data['date'].apply(get_month)
```

```python
data.head()
```

```python
data['cohortMonth']=data.groupby('user_id')['month_year'].transform('min')
```

```python
data.head(10)
```

```python
def get_date(df,column):
    year=df[column].dt.year
    month=df[column].dt.month
    day=df[column].dt.day
    return year,month,day
```

```python
date_year,date_month,_=get_date(data,'date')
```

```python
cohort_year,cohort_month,_=get_date(data,'cohortMonth')
```

```python
year_diff=date_year-cohort_year
```

```python
month_diff=date_month-cohort_month
```

```python
data['cohort_index']=year_diff*12+month_diff+1
```

```python
data.head()
```

```python
data['cohort_index'].unique()
```

```python
#1е задание для количества покупателей (user id)
data_customers=data.pivot_table(index='cohortMonth',
                      columns='cohort_index',
                      values='user_id',
                      aggfunc='nunique')
```

```python
data_customers
```

```python
first_column=data_customers.iloc[:,0]
```

```python
retention=data_customers.div(first_column,axis=0)
```

```python
retention.round(2)
```

```python
#2е задание для сумм их покупок
data_price=data.pivot_table(index='cohortMonth',
                      columns='cohort_index',
                      values='price',
                      aggfunc=sum)
```

```python
data_price
```

```python
first_column=data_price.iloc[:,0]
retention1=data_price.div(first_column,axis=0)
retention1.round(2)
```

```python
plt.figure(figsize=(11,9))
plt.title('когортный анализ суммы покупок-retention')
sns.heatmap(data=retention1,
           annot=True,
           fmt='.0%',
           vmin=0.0,
           vmax=0.5,
           cmap="YlGnBu")
plt.show()
```

```python
plt.figure(figsize=(11,9))
plt.title('когортный анализ количества клиентов-retention')
sns.heatmap(data=retention,
           annot=True,
           fmt='.0%',
           vmin=0.0,
           vmax=0.5,
           cmap="YlGnBu")
plt.show()
```

```python
#расчет среднего чека
data_check=data.pivot_table(index='cohortMonth',
                      columns='cohort_index',
                      values='price',
                      aggfunc='mean')
first_column=data_check.iloc[:,0]
retention2=data_check.div(first_column,axis=0)
retention2.round(2)
plt.figure(figsize=(11,9))
plt.title('когортный анализ среднего чека-retention')
sns.heatmap(data=retention2,
           annot=True,
           fmt='.0%',
           vmin=0.0,
           vmax=0.5,
           cmap="YlGnBu")
plt.show()
```

Количество новых пользователей растет волнообразно, в кагортах пришедших в четные месяцы идет уменьшение, в нечетные рост, 
возможно в нечетные месяцы действует какая-то акция, это подтверждает и поведение суммы покупок и среднего чека, 
они  растут в четные месяца.

То же самое касается и оттока новых пользовавтелей всех кагорт во 2й мес, у кагорт четных месяцев отток больше. 
Во 2м мес остается больший процент пользователей, пришедших в нечетные мес.

Наибольший отток пользователей во всех кагортах  происходит после 1го мес(до 70%), далее после 2го (до 24%), 
далее интенсивность оттока выравнивается.

Наблюдается наиболее интенсивный отток клиентов пришедших в 02.2020, с 7го месяца, учитывая что сумма покупок и средний чек растут,
то могу предположить, что их отпугнул рост цен.

Начиная с когорты 08.2020 динамика оттока клиентов перестает быть волнообразной и в каждой следующей кагорте отток чуть больше, 
либо равен оттоку в предыдущей кагорте.

Из-за снижения количества пользователей, снижается и стоимость покупок, что пытаются регулировать ростом цен, так как средний чек
растет.

В последнем месяце наблюдаются самые низкие значения суммы покупок и самые высокие значения среднего чека, при этом в 
снижении количества пользователей нет сильного скачка, но в 02.2021 пришло в 7 раз меньше клиентов, это показывает, то, что
месяц еще не закончился, вероятно прошло не более недели с его начала.
